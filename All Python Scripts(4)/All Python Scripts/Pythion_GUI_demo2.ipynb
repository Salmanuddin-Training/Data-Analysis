```python
#Python - Radiobutton widget
from tkinter import *

window = Tk()

var = IntVar()
var.set(2)

r1 = Radiobutton(window,text='Male',value=1, variable=var)
r2 = Radiobutton(window,text='Female',value=2,variable=var)

r1.place(x=100,y=50)
r2.place(x=180,y=50)

window.title('Radiobutton Demo..!')
window.geometry('500x200+30+20')
window.mainloop()
```

```python
#Python Combobox widget
from tkinter import *
from tkinter.ttk import Combobox

window =Tk()
var = StringVar()
var.set('Select....')

courses = ("C","C++","Python","Django")

cb = Combobox(window,values=courses, textvariable=var) 
cb.place(x=100,y=50)
window.title('Combobox Demo..!')
window.geometry('500x200+30+20')
window.mainloop()
```

```python
#Python Listbox widget
from tkinter import *

window =Tk()

fruits=("Apple","Banana",'Mango','Grapes','Papaya','Gauava')

lb = Listbox(window,height=4,selectmode='multiple')
for item in fruits:
    lb.insert(END,item)

lb.place(x=100,y=50)

window.title('Listbox Demo..!')
window.geometry('500x200+30+20')
window.mainloop()
```

```python
#Python- Event Handling - Button widget

from tkinter import *
from tkinter import messagebox

def myfun():
    print("Button Clicked.....!")
    #messagebox.showinfo('Message','Hi...Welcome to Python')
    #messagebox.showwarning('Message','Hi...Welcome to Python')
    #messagebox.showerror('Error','Hi...Welcome to Python')
    messagebox.askquestion('Message','Hi...Welcome to Python')

window = tkinter.Tk()

btn = Button(window,text='Click Me...!!!',command=myfun)
btn.place(x=200,y=100)

window.title('My First GUI in Python!!!')
window.geometry('500x200+30+20')
window.mainloop()
```

```python
#Python GUI Calculator

from tkinter import *

class MyCalculator:
    def add(self):
            self.t3.delete(0,'end')
            num1 = int(self.t1.get())
            num2 = int(self.t2.get())
            result=num1+num2
            self.t3.insert(END,str(result))
            
    def product(self):
            self.t3.delete(0,'end')
            num1 = int(self.t1.get())
            num2 = int(self.t2.get())
            result=num1*num2
            self.t3.insert(END,str(result))
    
    def __init__(self,win):
        self.lbl1=Label(win,text='Enter 1st no.')
        self.lbl2=Label(win,text='Enter 2nd no.')
        self.lbl3=Label(win,text='Result')
        
        self.t1=Entry(win,bd=4)
        self.t2=Entry(win,bd=4)
        self.t3=Entry(win,bd=4,bg='yellow',fg='red',font=('arial',15))
        
        self.btnADD=Button(win,text='ADD',bg='cyan',command=self.add)
        self.btnMUL=Button(win,text='MULTIPLY',bg='cyan',command=self.product)
        
        self.lbl1.place(x=100,y=50)
        self.t1.place(x=200,y=50)
        
        self.lbl2.place(x=100,y=100)
        self.t2.place(x=200,y=100)
        
        self.btnADD.place(x=200,y=150)
        self.btnMUL.place(x=300,y=150)
        
        self.lbl3.place(x=100,y=200)
        self.t3.place(x=200,y=200)
        
        

window =Tk()
MyCalculator(window)
window.title('My Python Calculator...!!!')
window.geometry('500x300+300+200')
window.mainloop()
```
