```python
#Python- MenuBar ,Menu ,Menuitems Demo......!
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

class MyProject:
    def addproduct(self):
        window = Tk()
        window.title('Product Details :-')
        window.geometry('300x400+500+200')
        window.mainloop()
    
    
    def __init__(self,win):
        #self.lbl1=Label(win,text="Welcome to Alpha Inventory System...!",fg='green',bg='yellow',font=('arial',15,'bold'))
        #self.lbl1.grid(row=0,column=1)
        
#============== MenuBar =======================================
        chooser = Menu()
        item1 = Menu(tearoff=0)
        item1.add_command(label='Add Product',command=self.addproduct)
        item1.add_command(label='Search/View Product')
        item1.add_command(label='Delete Product')
        item1.add_separator()
        item1.add_command(label='Help')
        item1.add_command(label='Exit')
        
        item2 = Menu(tearoff=0)
        item2.add_command(label='Add Customer')
        item2.add_command(label='Search/View Customer')
        item2.add_command(label='Delete Customer')
        item2.add_separator()
        item2.add_command(label='Help')
        item2.add_command(label='Exit')
        
        
        chooser.add_cascade(label='Product Master',menu=item1)
        chooser.add_cascade(label='Customer Master',menu=item2)
        win.config(menu=chooser)

if __name__ ==  '__main__':
    window = Tk()
    MyProject(window)
    window.title('Alpha Inventory System...!')
    window.geometry('530x465+500+200')
    window.mainloop()
```

```python
from tkinter import *
from tkinter.ttk import *
from tkinter import filedialog as fd
import MySQLdb

class MyProject:
    def convertToBinaryData(filename):
        #convert digital data to binary format
        with open(filename,'rb') as file:
            binarydata = file.read()
        return binarydata
    
    def browsePhoto(self):
        filename1 = fd.askopenfilename()
        self.photo.insert(END,filename1 )
    
    def browseResume(self):
        filename2 = fd.askopenfilename()
        self.resume.insert(END,filename2)
    
    def insertData(self):
        self.name=self.name.get() #    Williams
        self.photo=self.photo.get() #  E:/aa.jpg
        self.resume=self.resume.get()# E:/wil_resume.pdf
        
        try:
            db=MySQLdb.connect('localhost','root','root','python2_db')
            cursor = db.cursor()
            sqlquery = "insert into employee(name,photo,resume) values(%s,%s,%s)"
        
            empPicture = convertToBinaryData(self.photo)
            empResumefile = convertToBinaryData(self.resume)
        
            cursor.execute(sqlquery ,(self.name, empPicture,empResumefile))
            db.commit()
            print('Image and file inserted successfully into mysql database table...!')
            messagebox.showinfo('SUCCESS','Record has been saved successfully...!')
        
        except MySQLdb.Error as error:
            print("Failed to insert BLOb data :-",error)
        
    def __init__(self,win):
        self.lblname = Label(win,text='Name',font=('arial',12))
        self.lblname.place(x=20,y=50)
        self.name=Entry(win)
        self.name.place(x=100,y=50)
    
        self.lblphoto = Label(win,text='Photo',font=('arial',12))
        self.lblphoto.place(x=20,y=100)
        self.photo=Entry(win)
        self.photo.place(x=100,y=100)
        self.photobtn=Button(win,text='Select Photo',command=self.browsePhoto)
        self.photobtn.place(x=250,y=100)
        
        self.lblresume = Label(win,text='Resume',font=('arial',12))
        self.lblresume.place(x=20,y=150)
        self.resume=Entry(win)
        self.resume.place(x=100,y=150)
        self.resumebtn=Button(win,text='Select Resume',command=self.browseResume)
        self.resumebtn.place(x=250,y=150)
        
        self.submitbtn=Button(win,text='SUBMIT',command=self.insertData)
        self.submitbtn.place(x=100,y=200)

window = Tk()
window.title('Add employee....!')
MyProject(window)
window.geometry('530x465+500+200')
window.mainloop()
```
