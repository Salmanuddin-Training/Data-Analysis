```python
marks=[10,20,30,40,50]
sum=0
for m1 in marks:
    sum+=m1 #sum=sum+m1
avg = sum/len(marks)
print('\nAverage :',avg)
    
```

```python
#for loop iteration over dictionary
dict={"IND":"INDIA","CHN":"CHINA","GER":"GERMANY"}
for pair in dict.items():
    print(pair)
```

```python
#for loop iteration over dictionary
dict={"IND":"INDIA","CHN":"CHINA","GER":"GERMANY"}
for k,v in dict.items():
    print("Key : "+k+", Value = "+v)
```

```python
#for loop iteration over dictionary
dict={"IND":"INDIA","CHN":"CHINA","GER":"GERMANY"}
for k in dict.keys():
    print(k)
```

```python
#for loop iteration over dictionary
dict={"IND":"INDIA","CHN":"CHINA","GER":"GERMANY"}
for k in dict.keys():
    print(k,dict.get(k))
```

```python
#for loop iteration
num = int(input('Enter any no.'))
for i in range(1,11):
    print(num," * ",i," = ",(num*i))
```

```python
rows = 5
for i in range(0, rows):
    for j in range(0, i + 1):
        print("*", end=' ')

    print("\r")
```

```python
#else statement in for loop
for x in range(5):
    print(x+1)
else:
    print("else block in loop")
print("out of loop")
```

```python
#break statement 
for x in range(1,11):
    if x==3:
        continue
    if x==6:
        break
    print(x)
```

```python
#pass statement 
for num in range(1,6):
    if num==3:
        pass
    else:
        print("Num = {}",format(num))
```
