```python
import sys
garmentCol=[]
garmentplst =('trousers','suits','traditional wear','accessories','sports gear')
choice =0
class Garments:
    sales=0
    def __init__(self):
        self.__gID=0 
        self.__gtype=''
        self.__stock=0
        self.__price=0
        self.__orderqty=0
    
    def addGarments(self):
        self.__gID=int(input('Enter garment ID :'))
        while True:
            self.__gtype=input('Enter garment type-(trousers,suits,traditional wear,accessories,sports gear) :')
            if self.__gtype in garmentplst:
                break
            else:
                print("enter valid garment type......?")
                
        self.__stock=int(input('Enter garment stock :'))
        self.__price=int(input('Enter garment price :'))
        
    def showDetails(self):
        print("======= Garment Info :- =========")
        for prod in garmentCol:
            print("ID   : ",prod.__gID)
            print("Type : ",prod.__gtype)
            print("stock: ",prod.__stock)
            print("Price: ",prod.__price)
            print("===================================")
        
    def setOrder(self):
        pid=int(input('Enter garment ID :'))
        self.__orderqty=int(input('Enter order quantity :'))
        
        flag =True
        
        for prod in garmentCol:
            if pid == prod.__gID:
                Garments.sales=prod.__price * self.__orderqty
                flag=False
                break
        if flag:
            print('Invalid garment Id .....!')
        else:
            print('====================================')
            print("Payable amount is  : ",Garments.sales)
            print('====================================')

    
while True:
    g1=Garments()
    print("===== Fashionista Store===========")
    print("\t 1. Add Garment")
    print("\t 2. View Garments")
    print("\t 3. Search Garment")
    print("\t 4. Sale Order")
    print("\t 5. Exit")
    
    choice=int(input('Enter your choice between -(1 to 5):'))
    if choice==1:
        g1.addGarments()
        garmentCol.append(g1)
        continue
    elif choice==2:
        g1.showDetails()
    elif choice==3:
        pass
    elif choice==4:
        g1.setOrder()
    elif choice==5:
        sys.exit
    else:
        print('Invalid choice try again......!')
    

        
```
