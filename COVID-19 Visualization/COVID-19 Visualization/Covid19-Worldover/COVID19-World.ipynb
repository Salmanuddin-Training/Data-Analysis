```python
import requests
import csv
import json
import pandas as pd
from pandas import DataFrame as df
import os
import numpy as np
```

```python
#Confirmed cases
cw3_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')



# Death toll
cw3_data_dt = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')

#Recovered cases 
cw3_data_rec = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

#cw3_data_rec.head()
```

```python
last_col = cw3_data.columns[-1]
last_col_dt = cw3_data_dt.columns[-1]
last_col_rec = cw3_data_rec.columns[-1]

if last_col == last_col_dt == last_col_rec:
    col_ = last_col
col_
```

```python
cw3_data = cw3_data.groupby('Country/Region').sum()
cw3_data_dt = cw3_data_dt.groupby('Country/Region').sum()
cw3_data_rec = cw3_data_rec.groupby('Country/Region').sum()
cw3_data =  cw3_data.drop(cw3_data.columns[2:-1], axis =1)
cw3_data_rec.head()
```

```python
cw_data = cw3_data.reset_index()
cw_data_d = cw3_data_dt.reset_index()
cw_data_r = cw3_data_rec.reset_index()
```

```python
cw_data.rename(columns = {col_:'Confirmed_Cases', 'Country/Region' : 'Country'}, inplace = True)
cw_data.head()
```

```python
cw_data['Death_Toll'] = cw_data_d[col_].values
cw_data['Recovered_Cases'] = cw_data_r[col_].values
```

```python
tm_data = cw_data.drop(cw_data.columns[1:3], axis =1)
```

```python
tm_data_ = tm_data.assign(world = 'world')
tm_data_
```

```python
df = pd.DataFrame(tm_data_)
gfg_csv_data = df.to_csv('tableau_covid19_world.csv', index = True)
```
