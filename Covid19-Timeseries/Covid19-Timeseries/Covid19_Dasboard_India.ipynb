```python
# Dataset for all the Cities in India
import requests
import csv
import json
import pandas as pd
from pandas import DataFrame as df
import os
import numpy as np

JSON_URL = 'https://api.covid19india.org/v4/min/timeseries-MH.min.json'
URL = 'https://api.covid19india.org/v4/min/timeseries-'
EXT = '.min.json'
JSON_INDIA = 'https://api.covid19india.org/v4/min/timeseries.min.json'
```

```python
state_name = []


req_date = requests.get(JSON_URL)
req_India = requests.get(JSON_INDIA)

stateNames = df(req_India.json())
statesData = stateNames.T

for st in statesData.index:
     if not 'UN' in st:
         state_name.append(st)

unwanted_st = {'UN', 'TT', 'Other State', 'Other Region', 'Other', 'Unknown'} 
state_name = [ele for ele in state_name if ele not in unwanted_st]

    

dtcolNames =[]
covidData =[]
covidRec = []
covidDeath = []
otherData = []
tested = []
```

```python
dates = df(req_date.json()['MH']['districts']['Mumbai'])
for dt in dates.index:
    dtcolNames.append(dt + ",")
```

```python
for st in state_name:
    st_name = name_correction(st)
    #print ('statename: ' + st_name)
    dist_name  = []
  
    JSON_URL = URL + st + EXT
    req = requests.get(JSON_URL)    
    distNames = df(req.json())
    if 'districts' in distNames.index:
        distNames = df(req.json()[st]['districts'])
        distNames = distNames.T

        for dis in distNames.index:
            #print (st + " : " + dis)
            dist_name.append(dis)
 
  
        # items to be removed 
        unwanted_elem = {'Foreign Evacuees', 'Other State', 'Other Region', 'Others', 'Other', 'Unknown', 'Italians',
                         'Evacuees','BSF Camp', 'Airport Quarantine', 'Railway Quarantine', 'State Pool'} 
  
        dist_name = [ele for ele in dist_name if ele not in unwanted_elem]
            
        
                
        for dist  in dist_name:   
            
            i=0
            covidData.append('\n')
            
            
            if dist == st_name:
                dist_ = dist + " Region"
                #print(st_name + " : " + dist_)
                
            else:
                dist_ = dist  
            
            #print("final : " + dist_)
            covidData.append(st_name + "," + st + "," + dist_ + ",")
            
            covidRec.append('\n')
            covidRec.append(st_name + "," + st + "," + dist_ + ",")
            
            covidDeath.append('\n')
            covidDeath.append(st_name + "," + st + "," + dist_ + ",")
            
            otherData.append('\n')
            otherData.append(st_name + "," + st + "," + dist_ + ",")
            
            tested.append('\n')
            tested.append(st_name + "," + st + "," + dist_ + ",")
    
            covid = df(req.json()[st]['districts'][dist])
    
            
            for conf, dt in zip(covid.dates, covid.index):
                #print(dt)
                dt = dt + ","
                i+=1
                index = dtcolNames.index(dt) + 1
                if i!= index:
                    #print(st + ":" + dist + ":" + dt + ":  Ind: " + str(index) + ":" +  "i :" + str(i))
                    if i == 1:
                        for n in range(index-1):
                            covidData.append("0,")
                            covidRec.append("0,")
                            covidDeath.append("0,")
                            otherData.append("0,")
                            tested.append("No Data,")
                            
                    else:    
                        for n in range(index-i+1):
                            covidData.append("0,")
                            covidRec.append("0,")
                            covidDeath.append("0,")
                            otherData.append("0,")
                            tested.append("No Data,")
                            
                    i = index
                
               
                    
                for t in conf.keys():
                    if 'total' in t:
                        if 'confirmed' in (conf['total'].keys()):
                            covidData.append(str(conf['total']['confirmed']) + ",")
                        if not 'confirmed' in (conf['total'].keys()):
                            covidData.append("0,")
                    
                        if 'recovered' in (conf['total'].keys()):
                            covidRec.append(str(conf['total']['recovered']) + ",")
                        if not 'recovered' in (conf['total'].keys()):
                            covidRec.append("0,")
                    
                        if 'deceased' in (conf['total'].keys()):
                            covidDeath.append(str(conf['total']['deceased']) + ",")
                        if not 'deceased' in (conf['total'].keys()):
                            covidDeath.append("0,")
                        
                        if 'other' in (conf['total'].keys()):
                            otherData.append(str(conf['total']['other']) + ",")
                        if not 'other' in (conf['total'].keys()):
                            otherData.append("0,")
                        
                        if 'tested' in (conf['total'].keys()):
                            tested.append(str(conf['total']['tested']) + ",")
                        if not 'tested' in (conf['total'].keys()):
                            tested.append("No Data,")
                            
                            
                            
            if dt != dtcolNames[len(dtcolNames)-1]:
                diff = len(dtcolNames) - dtcolNames.index(dt)
                for n in range(diff):
                    covidData.append(covidData[len(covidData)-1])
                    covidDeath.append(covidDeath[len(covidDeath)-1])
                    covidRec.append(covidRec[len(covidRec)-1])
                    otherData.append(otherData[len(otherData)-1])
                    tested.append(tested[len(tested)-1])
                    
                
                
            
    else:
        print("State without sistrict's data : "+ st)
        
            
```

```python
with open('Indian_Cities_total_confirmed_cases.csv', 'w') as f:
    f.writelines("State,State Code,District,")
    f.writelines(dtcolNames)
    f.writelines(covidData)
    
with open('Indian_Cities_total_recovered_cases.csv', 'w') as f:
    f.writelines("State,State Code,District,")
    f.writelines(dtcolNames)
    f.writelines(covidRec)

with open('Indian_Cities_total_Death_toll.csv', 'w') as f:
    f.writelines("State,State Code,District,")
    f.writelines(dtcolNames)
    f.writelines(covidDeath)

with open('Indian_Cities_other_counts.csv', 'w') as f:
    f.writelines("State,State Code,District,")
    f.writelines(dtcolNames)
    f.writelines(otherData)

with open('Indian_Cities_tests_counts.csv', 'w') as f:
    f.writelines("State,State Code,District,")
    f.writelines(dtcolNames)
    f.writelines(tested)
```

## Analysis


### Confirmed cases per million:<br>How many cases out of every 10 lakh people in a particular region have tested positive for the virus?

```python
confirmed_per_mn = []

for i,j in zip(df_covid19['Population_Region'].values, df_covid19['Confirmed Cases'].values):
    if not 'NA' in str(i):
        pmn = (j/i)*1000000
        confirmed_per_mn.append(pmn)
        #print(k, ":", pmn)
    else:
        confirmed_per_mn.append('NA')

```

```python
df_covid19['Confirmed per million'] = confirmed_per_mn
#df_covid19['Confirmed per lakh'] = confirmed_per_lakh  # move left ny ine disgit 1mn : 25.18, 1 lakh : 2.51, 10k: 0.251
```

### Active Cases Ratio:<br>For every 100 confirmed cases, How many cases are currently infected?

```python
active_ratio = []
for i,j in zip(df_covid19['Active Cases'].values, df_covid19['Confirmed Cases'].values):
    if not j == 0:
        ar = (i/j)*100
        #print(k, ":", ar)
        active_ratio.append(ar)
    else:
        active_ratio.append(0)
    
```

```python
df_covid19['Active Ratio'] = active_ratio
```

### Recovered Cases Ratio:<br>For every 100 confirmed cases, How many have recovered from the virus?


```python
recovery_ratio = []
for i,j in zip(df_covid19['Recovered Cases'].values, df_covid19['Confirmed Cases'].values):
    if not j == 0:
        rer = (i/j)*100
        #print(k, ":", ar)
        recovery_ratio.append(rer)
    else:
        recovery_ratio.append(0)
```

```python
df_covid19['Recovery Ratio'] = recovery_ratio
```

### Case Fatality Ratio:<br>For every 100 confirmed cases, how many have have unfortunately passed away from the virus?


```python
case_fatality_ratio = []
for i,j in zip(df_covid19['Death Toll'].values, df_covid19['Confirmed Cases'].values):
    if not j == 0:
        cfr = (i/j)*100
        #print(k, ":", ar)
        case_fatality_ratio.append(cfr)
    else:
        case_fatality_ratio.append(0)
```

```python
df_covid19['Case Fatality Ratio'] = case_fatality_ratio
```

### Total Tests per million:<br>For every 10 lakh people in a particular region, How many samples were tested?
Always consider value of prior day as the data collection is completed for it and for current day collection is still in process.

```python
tests_per_mn = []

for i,j in zip(df_covid19['Tested'].values, df_covid19['Population_Region'].values):
    if not ('No Data' in str(i) or 'NA' in str(j)):
        tmn = (int(i)/j)*1000000
        tests_per_mn.append(tmn)
            #print(k, ":", pmn)
    else:
        tests_per_mn.append('NA')

```
